## 2. Data Preprocesssing 

```python
# 수치형 변수 리스트
numeric_fields = ['duration', 'src_bytes',
       'dst_bytes', 'wrong_fragment', 'urgent', 'hot',
       'num_failed_logins', 'num_compromised', 'root_shell',
       'su_attempted', 'num_root', 'num_file_creations', 'num_shells',
       'num_access_files', 'num_outbound_cmds', 'count', 'srv_count', 'serror_rate',
       'srv_serror_rate', 'rerror_rate', 'srv_rerror_rate', 'same_srv_rate',
       'diff_srv_rate', 'srv_diff_host_rate', 'dst_host_count',
       'dst_host_srv_count', 'dst_host_same_srv_rate',
       'dst_host_diff_srv_rate', 'dst_host_same_src_port_rate',
       'dst_host_srv_diff_host_rate', 'dst_host_serror_rate',
       'dst_host_srv_serror_rate', 'dst_host_rerror_rate',
       'dst_host_srv_rerror_rate']

# 부울형 변수 리스트 
binary_fields = ['land', 'logged_in', 'is_host_login','is_guest_login']

# 카테고리형 변수 리스트
category_fields = ['protocol_type1', 'protocol_type2', 'protocol_type3',
                   'service1', 'service2', 'service3', 'service4', 'service5',
                   'service6', 'service7', 'service8', 'service9', 'service10',
                   'service11', 'service12', 'service13', 'service14', 'service15',
                   'service16', 'service17', 'service18', 'service19', 'service20',
                   'service21', 'service22', 'service23', 'service24', 'service25',
                   'service26', 'service27', 'service28', 'service29', 'service30',
                   'service31', 'service32', 'service33', 'service34', 'service35',
                   'service36', 'service37', 'service38', 'service39', 'service40',
                   'service41', 'service42', 'service43', 'service44', 'service45',
                   'service46', 'service47', 'service48', 'service49', 'service50',
                   'service51', 'service52', 'service53', 'service54', 'service55',
                   'service56', 'service57', 'service58', 'service59', 'service60',
                   'service61', 'service62', 'service63', 'service64', 'service65',
                   'service66', 'service67', 'service68', 'service69', 'service70',
                   'flag1', 'flag2', 'flag3', 'flag4', 'flag5',
                   'flag6', 'flag7', 'flag8', 'flag9', 'flag10',
                   'flag11']

total_fields = numeric_fields + binary_fields + category_fields
```

```python
# 카테고리 형 변수 전처리: category_fields 
def category_fields_encoding(df):
    # 1) Integer Encoding
    from sklearn.preprocessing import LabelEncoder
    class_le = LabelEncoder()

    df['protocol_type'] = class_le.fit_transform(df['protocol_type'].values) 
    df['service'] = class_le.fit_transform(df['service'].values) 
    df['flag'] = class_le.fit_transform(df['flag'].values) 

    # 2) One-Hot Encoding
    from sklearn import preprocessing
    enc = preprocessing.OneHotEncoder()

    array_protocol_type_one_hot = enc.fit_transform(np.array(df['protocol_type']).reshape(-1, 1)).toarray()
    array_service_one_hot = enc.fit_transform(np.array(df['service']).reshape(-1, 1)).toarray()
    array_flag_one_hot = enc.fit_transform(np.array(df['flag']).reshape(-1, 1)).toarray()
    df_output = pd.DataFrame(np.column_stack((array_protocol_type_one_hot, array_service_one_hot, array_flag_one_hot)))
    df_output.columns = category_fields
    
    return df_output
```

```python
# 카테고리 형 변수 전처리: category_fields 
df_total[category_fields] = category_fields_encoding(df_total)
```

```python
# 수치형 변수 전처리: numeric_fields
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df_total[numeric_fields] = scaler.fit_transform(df_total[numeric_fields])
```

```python
# attack_name을 활용해 attack_class, attack_type 정의 하는 함수: add_target_column
# attack_class: normal, dos, u2r, r2l, probe
# attack_type: normal, abnormal 
def add_target_column(df):
    #df.at[normal_index, 'attack_class'] = 'normal'
    
    
    # normal_index 정의 
    normal_index = df.loc[(df['attack_name'] == 'normal')].index

    # dos index 정의: 'back', 'land', 'neptune', 'pod', 'smurf', 'teardrop'
    dos_index = df.loc[(df['attack_name'] == 'back') |
                       (df['attack_name'] == 'land') |
                       (df['attack_name'] == 'neptune') |
                       (df['attack_name'] == 'pod') |
                       (df['attack_name'] == 'smurf') |
                       (df['attack_name'] == 'teardrop') |                
                       (df['attack_name'] == 'apache2') |
                       (df['attack_name'] == 'udpstorm') |
                       (df['attack_name'] == 'processtable') |
                       (df['attack_name'] == 'worm')].index

    # u2r index 정의: 'buffer_overflow', 'loadmodule', 'perl', 'rootkit'
    u2r_index = df.loc[(df['attack_name'] == 'buffer_overflow') |
                       (df['attack_name'] == 'loadmodule') |
                       (df['attack_name'] == 'rootkit') |
                       (df['attack_name'] == 'perl') |
                       (df['attack_name'] == 'sqlattack') |
                       (df['attack_name'] == 'xterm') |
                       (df['attack_name'] == 'ps')].index

    # r2l index 정의: 'ftp_write', 'guess_passwd', 'imap', 'multihop', 'phf', 'spy', 'warezclient', 'warezmaster'
    r2l_index = df.loc[(df['attack_name'] == 'guess_passwd') |
                       (df['attack_name'] == 'ftp_write') |
                       (df['attack_name'] == 'imap') |
                       (df['attack_name'] == 'phf') |
                       (df['attack_name'] == 'multihop') |
                       (df['attack_name'] == 'warezmaster') |
                       (df['attack_name'] == 'warezclient') |
                       (df['attack_name'] == 'warezclient') |
                       (df['attack_name'] == 'spy') |
                       (df['attack_name'] == 'xlock') |
                       (df['attack_name'] == 'xsnoop') |
                       (df['attack_name'] == 'snmpguess') |
                       (df['attack_name'] == 'snmpgetattack') |
                       (df['attack_name'] == 'httptunnel') |
                       (df['attack_name'] == 'sendmail') |
                       (df['attack_name'] == 'named')].index

    # probe index 정의: 'ipsweep', 'nmap', 'portsweep', 'satan'
    probe_index = df.loc[(df['attack_name'] == 'ipsweep') |
                         (df['attack_name'] == 'nmap') |
                         (df['attack_name'] == 'portsweep') |
                         (df['attack_name'] == 'satan') |
                         (df['attack_name'] == 'mscan') |
                         (df['attack_name'] == 'saint')].index
    
    # etc_index 정의: 'mailbolmb'
    etc_index = df.loc[(df['attack_name'] == 'mailbomb')].index
    
    # abnormal index 정의: dos_index, u2r_index, r2l_index, probe_index
    abnormal_index = np.append(np.array(dos_index), np.array(u2r_index))  # abnormal_index에 dos_index, u2r_index 추가
    abnormal_index = np.append(abnormal_index, np.array(r2l_index))       # abnormal_index에 r2l_index 추가 
    abnormal_index = np.append(abnormal_index, np.array(probe_index))     # abnormal_index에 probe_index 추가 
    abnormal_index = np.append(abnormal_index, np.array(etc_index))       # abnormal_index에 etc_index 추가 
    abnormal_index = pd.Index(abnormal_index)
    abnormal_index
    
    
    # attack_class 정의
    df.at[normal_index, 'attack_class'] = 'normal'
    df.at[dos_index, 'attack_class'] = 'dos'
    df.at[u2r_index, 'attack_class'] = 'u2r'
    df.at[r2l_index, 'attack_class'] = 'r2l'
    df.at[probe_index, 'attack_class'] = 'probe'
    df.at[etc_index, 'attack_class'] = 'etc'
    
    # attack_type 정의
    df.at[normal_index, 'attack_type'] = 'normal'
    df.at[abnormal_index, 'attack_type'] = 'abnormal'
    
    return df
```

```python
# new_df = df_total에 attack_class, attack_type 컬럼 추가  
df_new = add_target_column(df_total)
df_new[['attack_name', 'attack_class', 'attack_type']].describe()
```

# 3. Splitting & Storing Data
1. Training Data = Total Data의 85%
    + Train Data 1: Normal 99%, Abnormal 1%
    + Train Data 2: Normal 95%, Abnormal 5%
    + Train Data 3: Normal 90%, Abnormal 10%
2. Test Data = Total Data의 15%

```python
# data shuffling
from sklearn.utils import shuffle
df_new = shuffle(df_new, random_state = 0)
```

```python
# Total Data 쪼개기: Train 85%, Test 15%
train_ratio = 0.85
test_ratio = 0.15 

total_num = len(df_new.index)
train_num = int(train_ratio*total_num)
test_num = total_num

df_train = df_new.iloc[0:train_num] # normal_training_data: 0 ~ nomral_train_num
df_test = df_new.iloc[train_num:test_num] # normal_test_data: normal_validation_num ~ normal_test_num 
```

```python
print(len(df_train.index),len(df_test.index))
```

```python
# Train Data의 normal/abnormal 쪼개기
df_train_normal = df_train[df_train['attack_type'] == 'normal']
df_train_abnormal = df_train[df_train['attack_type'] == 'abnormal']

# shuffling
df_train_normal = shuffle(df_train_normal, random_state = 0)
df_train_abnormal = shuffle(df_train_abnormal, random_state = 0)
```

```python
# resampling abnormal data 
normal_train1_normal_ratio = 0.99
normal_train2_normal_ratio = 0.95
normal_train3_normal_ratio = 0.90
#normal_test_abnormal_ratio = 1 - normal_train_normal_ratio


abnormal_train1_num = int((1/normal_train1_normal_ratio - 1)* len(df_train_normal.index))
abnormal_train2_num = int((1/normal_train2_normal_ratio - 1)* len(df_train_normal.index))
abnormal_train3_num = int((1/normal_train3_normal_ratio - 1)* len(df_train_normal.index))

df_train1_abnormal = df_train_abnormal.iloc[0:abnormal_train1_num]
df_train2_abnormal = df_train_abnormal.iloc[0:abnormal_train2_num]
df_train3_abnormal = df_train_abnormal.iloc[0:abnormal_train3_num]
```

```python
# df_train1 = 99% 정상데이터 + 1% 이상데이터 
df_train1 = df_train_normal.append(df_train1_abnormal,ignore_index=True)

# df_train2 = 95% 정상데이터 + 5% 이상데이터 
df_train2 = df_train_normal.append(df_train2_abnormal,ignore_index=True)

# df_train3 = 90% 정상데이터 + 10% 이상데이터 
df_train3 = df_train_normal.append(df_train3_abnormal,ignore_index=True)
```

```python
# shuffling
df_train1 = shuffle(df_train1, random_state = 0)
df_train2 = shuffle(df_train2, random_state = 0)
df_train3 = shuffle(df_train3, random_state = 0)

df_test = shuffle(df_test, random_state = 0)


# 데이터 저장 
df_train1.to_csv('data/df_train1(preprocessed).csv', index=False)
df_train2.to_csv('data/df_train2(preprocessed).csv', index=False)
df_train3.to_csv('data/df_train3(preprocessed).csv', index=False)

df_test.to_csv('data/df_test(preprocessed).csv', index=False)
```

# 4. Exploring Data

```python
# attack type count를 계산하는 함수 calculate_attack_type 정의
def calculate_attack_type(df):
    normal_count = len(df[df['attack_type'] == 'normal'])
    abnormal_count = len(df[df['attack_type'] == 'abnormal'])

    print("normal_count: %d, \nabnormal_count: %d\n"
           %(normal_count, abnormal_count))

# attack class count를 계산하는 함수 calculate_attack_class 정의
def calculate_attack_class(df):
    dos_class = len(df[df['attack_class'] == 'dos'])
    etc_class = len(df[df['attack_class'] == 'etc'])
    normal_class = len(df[df['attack_class'] == 'normal'])
    probe_class = len(df[df['attack_class'] == 'probe'])
    r2l_class = len(df[df['attack_class'] == 'r2l'])
    u2r_class = len(df[df['attack_class'] == 'u2r'])
    
    print("dos_class: %d, \netc_class: %d, \nnormal_class: %d, \nprobe_class: %d, \nr2l_class: %d, \nu2r_class: %d\n"
           %(dos_class, etc_class, normal_class, probe_class, r2l_class, u2r_class))
```

```python
# data import
df_train1 = pd.read_csv('data/df_train1(preprocessed).csv')
df_train2 = pd.read_csv('data/df_train2(preprocessed).csv')
df_train3 = pd.read_csv('data/df_train3(preprocessed).csv')
df_test = pd.read_csv('data/df_test(preprocessed).csv')
```

```python
# df_train의 통계량 계산 
print("\n***** df_train1 statistics *****\n")
calculate_attack_type(df_train1)
calculate_attack_class(df_train1)
print("*******************************\n")

print("\n***** df_train2 statistics *****\n")
calculate_attack_type(df_train2)
calculate_attack_class(df_train2)
print("*******************************\n")

print("\n***** df_train3 statistics *****\n")
calculate_attack_type(df_train3)
calculate_attack_class(df_train3)
print("*******************************\n")

# df_test의 통계량 계산 
print("\n***** df_test statistics *****\n")
calculate_attack_type(df_test)
calculate_attack_class(df_test)
print("*******************************\n")
```
